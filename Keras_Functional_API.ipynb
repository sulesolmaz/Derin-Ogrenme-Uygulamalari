##Kütüphane kurulumları ve ağların oluşturulması

```
from keras import Input, layers
```

```
#Bir modülden çıkan bir tensorü başka bir modülün girdisi yada iki sonraki modülün girdisi  haline getirebiliriz. 
input_tensor = Input((32,))#32 kanalı olana bir tensor
dense = layers.Dense(32, activation = 'relu')#bu tensörü dense layerda işleyebiliriz.
output_tensor = dense(input_tensor)#dense ile input tensörü aldık.
```

```
from keras.models import Sequential, Model
```

##Sequential Model

```
seq_model = Sequential()
seq_model.add(layers.Dense(32, activation='relu', input_shape= (64,))) 
seq_model.add(layers.Dense(32, activation='relu'))
seq_model.add(layers.Dense(10, activation='sigmoid'))

seq_model.summary()
```

##functional Model

```
input_tensor = Input(shape= (64,))

x = layers.Dense(32, activation = 'relu')(input_tensor)

x = layers.Dense(32, activation='relu')(x)

output_tensor = layers.Dense(10, activation = 'softmax')(x)

model = Model(input_tensor, output_tensor)#Aradakileri atlamış olduk sadece input ve outputu aldık.

model.summary()
```

##Modelin Derlenmesi

```
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy')
```

##Eğitim için rastgele bir küme oluşturmak

```
import numpy as np
```

```
x_train = np.random.random((1000, 64))
y_train = np.random.random((1000, 10))
```

##Modelin Eğitilmesi

```
model.fit(x_train, y_train, epochs = 10, batch_size=128)

score = model.evaluate(x_train, y_train)
```
